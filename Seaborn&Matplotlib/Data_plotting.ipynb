In this program we use data provided by our instructor and plot it in different types of graphs by using
the "Seaborn" and "Matplotlib" library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Excercise A
tips = pd.read_csv('workerstips.csv')

sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.title("Tips Scatter Plot")
plt.show()
```

```python
#Exercise B
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'size', data = tips, sizes = (10,300))
plt.title("Smoker and Non-Smoker Scatter Plot")
plt.show()
```

```python
#Excercise C
sns.barplot(x = 'day', y = 'tip', data = tips)
plt.title("Average Tips Bar Plot")
plt.show()
```

```python
#Exercise D
sns.boxplot(x = 'day', y ='tip', hue = 'time',data = tips)
plt.show()
```

```python
#Excercise E
flights = pd.read_csv('flightsData.csv')
sns.lineplot(x = 'year', y = 'passengers', data = flights)
plt.title("Flights Line Plot")
plt.show()
```

```python
#Excercise F
boat = pd.read_csv('titanic.csv')
boat['Pclass'] = boat['Pclass'].replace([1,2,3],['First','Second','Third'])

sns.catplot(x = 'Pclass', y = 'PassengerId', col = 'Survived', hue = 'Sex',data = boat, kind = "bar")
plt.show()
```
